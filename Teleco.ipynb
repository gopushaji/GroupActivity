```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Telco-Customer-Churn.csv')
data.columns
```

```python
data.head()
```

# 1.Compare Churn count wrt Gender

```python
churn_yes= data[data['Churn']=='Yes']
sns.countplot(x='gender', data=data, hue='Churn')
```

# 2 Find out how many female senior citizens are there in the dataset

```python
female= data[data['gender']=='Female']
```

```python
sns.countplot(x='SeniorCitizen', data=female)
plt.title('Distribution of Female Senior Citizen',fontsize=15)
plt.xlabel('Senior Citizen')
plt.ylabel('Number')
plt.xticks([0,1],['No','Yes'])
```

# 3.Compare Tenure with Total Charges

```python
sns.scatterplot(x='tenure',y='TotalCharges', data =data)
```

# 4.Find out which contract is preffered by the senior citizen

```python
senior_data=data[data['SeniorCitizen']==1]
plt.hist('Contract', data=senior_data)
```
