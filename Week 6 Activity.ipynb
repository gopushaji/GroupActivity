```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('G:\DSA\Activity\week6.csv')
```

```python

data.columns=['country', 'Region', 'Surface area (km2)',
       'Population in thousands (2017)', 'GDP per capita (current US$)',
       'Agricultural production index',
       'International trade: Imports (million US$)']
data.head()
```

# Question 1

```python
print('No. of countries in the list :', data['country'].nunique())
```

# Question 2

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
sns.boxplot(x='Region',y='GDP per capita (current US$)',data=data)
plt.xticks(rotation=90)

plt.subplot(1,2,2)
sns.boxplot(x='Region',y='International trade: Imports (million US$)',data=data)
plt.xticks(rotation=90)

plt.show
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GDP per capita (current US$)',y='International trade: Imports (million US$)',hue='Region',data=data)
plt.show
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
sns.scatterplot(x='Population in thousands (2017)',y='International trade: Imports (million US$)',data=data)

plt.subplot(1,2,2)
sns.scatterplot(x='Population in thousands (2017)',y='GDP per capita (current US$)',data=data)

plt.show
```

```python
sns.countplot(x='Region',data=data)
plt.xticks(rotation=90)
plt.show
```

# Question 4a

```python
reg_gdp=data.groupby('Region')[['GDP per capita (current US$)']].mean()
reg_gdp.nlargest(1,'GDP per capita (current US$)')
```

# Question 4b

```python
reg_imp=data.groupby('Region')[['International trade: Imports (million US$)']].mean()
reg_imp.nlargest(1,'International trade: Imports (million US$)')
```
